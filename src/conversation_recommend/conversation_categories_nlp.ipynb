```python
categories = df_user_interest['name'].unique()
category_list = []
for i in categories:
  category_list.append(i)
category_list
```

```python
#to download neccessary tokenisers, vocab etc for nlp
from transformers import pipeline

pipe = pipeline(model="facebook/bart-large-mnli")
```

```python
df_conversation['text'] = df_conversation['title'] + ' ' + df_conversation['caption']
```

```python
categories_captions = []
count = 0
for i in df_conversation['text']:
  if count > 5:
    break
  count += 1
  output = pipe(i,
      candidate_labels=category_list,
  )
  categories_captions.append(output['labels'][0])
categories_captions
```

```python
categories_captions = []

# Process each input using the reusable pipeline
for text in df_conversation['text']:
    result = pipe(
        text,
        candidate_labels=category_list,
    )
    categories_captions.append(result['labels'][0])
```

```python
df_conversation['categories'] = categories_captions
```
